1️⃣ Prints every seat with its position
Example output :

Row 1 Seat 3: Leah

2️⃣ Counts the total number of students
3️⃣ Finds and prints the longest student name in the entire seating grid
note : If there is a tie, you may print any one of the longest names.

Requirements
Use nested loops
Do not use max()

Your program must work for any grid size

```python
def list_seats(seats):
  count = 0
  longest = 0
  for i in range(len(seats)):
    for j in range(len(seats[i])):
      count +=1
      length = len(seats[i][j])
      if length >= longest:
        longest = len(seats[i][j])
        name = seats[i][j]
      print("Row", i+1, "Seat" , j+1, ":", seats[i][j])
    print()
  print("Count of students:",count)
  print("Longest name:",name)





list_seats([["Avi","Mira","Leah"],["Noah","Sara","Ben"],["Liam","Emma","Chana"]])
```

Write a function:

def digit_counts(n):
Input: a non-negative integer

Return a dictionary counting how many times each digit appears.

Example

digit_counts(455241)

{4:2, 5:2, 2:1, 1:1}
Requirements:

Convert digits properly

Use a dictionary

You may use any data type to assist you in this process, and any method associated with that method.
exceptions : count
Do not use collections.Counter

```python
def digit_counts(n):
  my_dict = {}
  for i in n:
    if i in my_dict:
      my_dict[i] +=1
    else:
      my_dict[i] = 1
  return my_dict
digit_counts("455241")
```

Each student record is stored as a tuple:

(name, assignment_score, exam_score)
Example:

("Avi",82,91)
Write the function:

def generate_record():
that randomly creates one student record.

Names may come from a list such as:

["Avi","Leah","Mira","Noah","Sara"]
Scores should be random integers between 60 and 100.

Then write:

def build_class(n):
that generates n records and returns a list of tuples.

Then write code that returns and prints:

1️⃣ all student records
2️⃣ the average exam score
3️⃣ the highest assignment score

Requirements:

Use tuples
Use loops
Look at cheatsheet, random library is provided in full with detailed usage.
Do not use max() or sum()

```python
import random
def generate_record(names):
  new_list = []
  for i in names:
    assignment_score = random.randint(60,100)
    exam_score = random.randint(60,100)
    new_list.append((i,assignment_score,exam_score))

  return new_list

def build_class(n, new_list):
  highest = new_list[0][1]
  total = 0
  for j in new_list:
    total+=j[2]
    if j[1] > highest:
      highest = j[1]

  average = total / len(new_list)
  n_students = []
  for i in range(len(new_list)):
    if i < n:
      n_students.append(new_list[i])
  return n_students, average, highest



new_list = generate_record(["Avi","Leah","Mira","Noah","Sara"])
n_students, average, highest = build_class(2, new_list)

print("Student Records:")
for i in new_list:
  print(i)
print()
print("Average Exam Scores:", average)
print("Highest Assignment Score:", highest)

print()
print("Generated n records:", n_students)
```

Write the function

def compress_string(s):
The function compresses repeated characters, and returns a string

Example

aaabbcaaa
becomes

a3b2c1a3
Rules:


Build the result string step by step
Case sensitive
No regex
You may assume that the input is not empty (you do not have to check for this case)

```python
def compress_string(s):
  string = ""
  num = 0
  letter = s[0]
  for i in s:
    if letter == i:
      num +=1
      n = str(num)
      if i not in string:
        string += i + n
      else:
        string += i + n
    else:
      num = 1
      n = str(num)
      string += i + n
    letter = i
  return string



print(compress_string("aaabbcaaa"))
```
